```python
#5 Chuẩn bị dữ liệu
#5.1 Khởi tạo bộ dữ liệu chuẩn hóa
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
#5.2 Chuẩn hóa huấn luyện
X_train_scaled = scaler.fit_transform(X_train)
#5.3 Chuẩn hóa kiểm tra
X_test_scaled = scaler.transform(X_test)


```

```python
#6 Ứng dụng mô hình hồi quy tuyến tính
#6.1 Khởi tạo
from sklearn.linear_model import LinearRegression

model = LinearRegression()
#6.2 Huấn luyện
model.fit(X_train_scaled, y_train)
#6.3 Dự đoán trên tập kiểm tra
y_pred = model.predict(X_test_scaled)
```

```python
#7. Đánh giá mô hình 
#Tính MSE và R^22
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'MSE: {mse}')
print(f'R^2: {r2}')


```

```python
#Bài tập về nhà
import pandas as pd
import numpy as np

# Tải dữ liệu (giả sử dữ liệu của bạn nằm trong file 'howlongwelive.csv')
df = pd.read_csv('howlongwelive.csv')

# 2.1 Số lượng cột và hàng
print(f"Số lượng hàng: {df.shape[0]}")
print(f"Số lượng cột: {df.shape[1]}")

# 2.2 Danh sách tên cột
print(f"Danh sách tên cột: {df.columns.tolist()}")

```

```python
# 2.3 Thử dùng df["status"]
print(df["Status"].head())

# 2.4 Số lượng nước đang phát triển và phát triển
print(df["Status"].value_counts())

# 2.5 Min, max, mean, median của alcohol
print(f"Min alcohol: {df['Alcohol'].min()}")
print(f"Max alcohol: {df['Alcohol'].max()}")
print(f"Mean alcohol: {df['Alcohol'].mean()}")
print(f"Median alcohol: {df['Alcohol'].median()}")

# 2.6 Bài 1: Đếm những hàng có rượu nhiều hơn mức trung bình và liệt kê những quốc gia độc đáo
mean_alcohol = df['Alcohol'].mean()
high_alcohol_countries = df[df['Alcohol'] > mean_alcohol]['Country'].unique()
print(f"Các quốc gia có lượng rượu cao hơn mức trung bình: {high_alcohol_countries}")
print(f"Số lượng quốc gia: {len(high_alcohol_countries)}")

# 2.6 Bài 2: Lấy danh sách quốc gia có trình độ học vấn cao hơn mức trung bình và GDP cao hơn mức trung bình
mean_schooling = df['Schooling'].mean()
mean_gdp = df['GDP'].mean()

high_schooling_gdp_countries = df[(df['Schooling'] > mean_schooling) & (df['GDP'] > mean_gdp)]['Country'].unique()
print(f"Các quốc gia có trình độ học vấn và GDP cao hơn mức trung bình: {high_schooling_gdp_countries}")
print(f"Số lượng quốc gia: {len(high_schooling_gdp_countries)}")
```
